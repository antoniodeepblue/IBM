### Data Preprocessing


#### One Hot Encoding


First, we need to perform one hot encoding to convert categorical variables to binary variables.


Primero, necesitamos realizar una codificación en caliente para convertir las variables categóricas en variables binarias.

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

Next, we replace the values of the 'RainTomorrow' column changing them from a categorical column to a binary column. We do not use the `get_dummies` method because we would end up with two columns for 'RainTomorrow' and we do not want, since 'RainTomorrow' is our target.


A continuación, reemplazamos los valores de la columna 'RainTomorrow' cambiándolos de una columna categórica a una columna binaria. No utilizamos el get_dummies método porque terminaríamos con dos columnas para 'RainTomorrow' y no queremos, ya que 'RainTomorrow' es nuestro objetivo.

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

### Training Data and Test Data


Now, we set our 'features' or x values and our Y or target variable.


Ahora, establecemos nuestras 'características' o valores x y nuestra variable Y o objetivo.

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
features
```

### Linear Regression


#### Q1) Use the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `10`.


Utilice la train_test_split función para dividir los marcos de datos featuresy Y con un test_size de 0.2 y el random_stat econjunto en 10.

```python
#Enter Your Code and Execute
X_train, X_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
print('X_train.shape=', X_train.shape, 'Y_train.shape=', y_train.shape)
print('X_test.shape=', X_test.shape, 'Y_test.shape=', y_test.shape)
```

#### Q2) Create and train a Linear Regression model called LinearReg using the training data (`x_train`, `y_train`).

```python
#Enter Your Code and Execute
```

```python
from sklearn import linear_model
```

```python
LinearReg = linear_model.LinearRegression()


LinearReg.fit(X_train, y_train)
```

```python
# The coefficients
print ('Coefficients: ', LinearReg.coef_)
print ('Intercept: ', LinearReg.intercept_)
```

#### Q3) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


Ahora use el predictmétodo en los datos de prueba ( x_test) y guárdelo en la matriz predictions.

```python
#Enter Your Code and Execute
```

```python
predictions = LinearReg.predict(X_test)
```

```python
predictions
```

#### Q4) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


Utilizando predictions y el y_test marco de datos calcule el valor de cada métrica utilizando la función adecuada.

```python
#Enter Your Code and Execute

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# y_pred = predictions
y_pred = LinearReg.predict(X_test)

# Calculate the metrics
LinearRegression_MAE = mean_absolute_error(y_test, y_pred)
LinearRegression_MSE = mean_squared_error(y_test, y_pred)
LinearRegression_R2 = r2_score(y_test, y_pred)
```

Importamos las funciones necesarias: Importamos `mean_absolute_error, mean_squared_error, y r2_score` desde sklearn.metricspara calcular las métricas deseadas.

Realizar predicciones: se utiliza LinearReg.predict(X_test) para predecir los valores objetivo de los datos de prueba (X_test). Las predicciones se almacenan en la y_pred variable.

Calcular métricas: cada función toma los valores objetivo reales (y_test) y los valores predichos (y_pred) como argumentos.

`mean_absolute_error`:Calcula la diferencia absoluta promedio entre los valores reales y previstos.

`mean_squared_error`:Calcula la diferencia cuadrática promedio entre los valores reales y previstos.

`r2_score`: Calcula la puntuación R-cuadrado, que es una medida de qué tan bien la línea de regresión se ajusta a los datos. Los valores más cercanos a 1 indican un mejor ajuste.

La interpretación de estas métricas puede depender de la escala de los datos. Por ejemplo, un MAE pequeño puede ser bueno para un conjunto de datos, pero no necesariamente para otro. Es útil comparar las métricas con el contexto y las expectativas de su problema específico.

```python
print("Linear Regression - Mean Absolute Error:", LinearRegression_MAE)
print("Linear Regression - Mean Squared Error:", LinearRegression_MSE)
print("Linear Regression - R-squared:", LinearRegression_R2)
```

#### Q5) Show the MAE, MSE, and R2 in a tabular format using data frame for the linear model.


Muestre el MAE, MSE y R2 en un formato tabular utilizando el marco de datos para el modelo lineal.

```python
#Enter Your Code and Execute

# Create a dictionary to store the metrics
metrics = {
    'Metric': ['Mean Absolute Error (MAE)', 'Mean Squared Error (MSE)', 'R-squared (R²)'],
    'Value': [LinearRegression_MAE, LinearRegression_MSE, LinearRegression_R2]
}

# Create a DataFrame from the dictionary
Report = pd.DataFrame(metrics)

# Print the DataFrame
print(Report)
```
