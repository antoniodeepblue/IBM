## Define Graphing Function


In this section, we define the function `make_graph`. **You don't have to know how the function works, you should only care about the inputs. It takes a dataframe with stock data (dataframe must contain Date and Close columns), a dataframe with revenue data (dataframe must contain Date and Revenue columns), and the name of the stock.**


````python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    stock_data_specific = stock_data[stock_data.Date <= '2021--06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date), y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date), y=revenue_data_specific.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

tesla_data["Date"] = pd.to_datetime(tesla_data["Date"]).dt.date
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"]).dt.date

def make_graph(stock_data, revenue_data, company_name):
    """
    Crea un gráfico que compara el precio de las acciones y los ingresos de una empresa.

    Args:
        stock_data: DataFrame con los datos de las acciones.
        revenue_data: DataFrame con los datos de los ingresos.
        company_name: Nombre de la empresa.
    """

    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Gráfico de precios de las acciones
    ax1.plot(stock_data['Date'], stock_data['Close'], color='blue', label='Precio de las acciones')
    ax1.set_xlabel('Fecha')
    ax1.set_ylabel('Precio de las acciones (USD)', color='blue')

    # Gráfico de ingresos
    ax2 = ax1.twinx()
    ax2.plot(revenue_data['Date'], revenue_data['Revenue'], color='green', label='Ingresos')
    ax2.set_ylabel('Ingresos (USD)', color='green')

    # Título y leyenda
    plt.title(f'Precio de las acciones y ingresos de {company_name} hasta junio de 2021')
    plt.legend()

    # Formato de fecha
    plt.gcf().autofmt_xdate()

    # Mostrar el gráfico
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Llamar a la función
make_graph(tesla_data, tesla_revenue, 'Tesla')
````

````python
# Obtener datos bursátiles de GameStop
gamestop = yf.Ticker("GME")
datos_bolsa_gamestop = gamestop.history(period="max")
datos_bolsa_gamestop.reset_index(inplace=True)

# Asegurar que la columna "Date" solo contenga información de fecha
datos_bolsa_gamestop["Date"] = pd.to_datetime(datos_bolsa_gamestop["Date"]).dt.date

# Web scraping para datos de ingresos de GameStop
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
datos_html_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(datos_html_gme, 'html.parser')

ingresos_gme = pd.DataFrame(columns=["Date", "Revenue"])

# Analizar los datos de ingresos
for tabla in soup_gme.find_all('table'):
  if tabla.find('th').getText().startswith("GameStop Quarterly Revenue"):
    for fila in tabla.find("tbody").find_all("tr"):
      columna = fila.find_all("td")
      if len(columna) != 2: continue
      Date = columna[0].text
      Revenue = columna[1].text.replace("$", "").replace(",", "")
      nueva_fila = pd.DataFrame({"Date": [Date], "Revenue": [Revenue]})
      ingresos_gme = pd.concat([ingresos_gme, nueva_fila], ignore_index=True)

# Convertir la columna "Fecha" a datetime y eliminar información de zona horaria
ingresos_gme["Date"] = pd.to_datetime(ingresos_gme["Date"]).dt.date

# Combinar los datos bursátiles y de ingresos por fecha
datos_combinados = pd.merge(datos_bolsa_gamestop, ingresos_gme, on="Date", how="left")

# Llenar valores faltantes en "Ingresos"
datos_combinados["Revenue"] = pd.to_numeric(datos_combinados["Revenue"], errors='coerce')
datos_combinados["Revenue"].fillna(method="ffill", inplace=True)

# Función para crear el gráfico
def crear_grafico(datos_bolsa, nombre_empresa):
  fig, ax1 = plt.subplots(figsize=(12,6))

  # Gráfico del precio de las acciones
  ax1.plot(datos_bolsa["Date"], datos_bolsa["Close"], color='red', label='Precio de las acciones')
  ax1.set_xlabel('Fecha')
  ax1.set_ylabel('Precio de las acciones (USD)', color='red')
  ax1.tick_params(axis='y', labelcolor='red')

  # Gráfico de ingresos
  ax2 = ax1.twinx()
  ax2.plot(datos_bolsa["Date"], datos_bolsa["Revenue"], color='blue', label='Ingresos')
  ax2.set_ylabel('Ingresos (USD)', color='blue')
  ax2.tick_params(axis='y', labelcolor='blue')

  # Título del gráfico
  plt.title(f'Precio de las acciones e ingresos de {nombre_empresa} a lo largo del tiempo')

  # Leyenda y formato
  plt.legend()
  plt.gcf().autofmt_xdate()
  plt.grid(True)
  plt.tight_layout()
  plt.show()

# Llamar a la función para el gráfico de GameStop
crear_grafico(datos_combinados, 'GameStop')
````

````python
# Get GameStop stock data
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="max")
gamestop_data.reset_index(inplace=True)

# Ensure Date column contains only date information
gamestop_data["Date"] = pd.to_datetime(gamestop_data["Date"]).dt.date

# Web scraping for GameStop revenue data
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(html_data_gme, 'html.parser')

gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Parse the revenue data
for table in soup_gme.find_all('table'):
    if table.find('th').getText().startswith("GameStop Quarterly Revenue"):
        for row in table.find("tbody").find_all("tr"):
            col = row.find_all("td")
            if len(col) != 2: continue
            Date = col[0].text
            Revenue = col[1].text.replace("$", "").replace(",", "")
            new_row = pd.DataFrame({"Date": [Date], "Revenue": [Revenue]})
            gme_revenue = pd.concat([gme_revenue, new_row], ignore_index=True)

# Convert the Date column to datetime and remove timezone info
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"]).dt.date

# Merge the stock data and revenue data based on Date
combined_data = pd.merge(gamestop_data, gme_revenue, on="Date", how="left")

# Fill missing revenue values
combined_data["Revenue"] = pd.to_numeric(combined_data["Revenue"], errors='coerce')
combined_data["Revenue"].fillna(method="ffill", inplace=True)

# Function to plot the stock price and revenue
def make_graph(stock_data, company_name):
    fig, ax1 = plt.subplots(figsize=(10,6))

    # Plot the stock price
    ax1.plot(stock_data["Date"], stock_data["Close"], color='blue', label='Stock Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price (USD)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Create second y-axis for revenue
    ax2 = ax1.twinx()
    ax2.plot(stock_data["Date"], stock_data["Revenue"], color='green', label='Revenue')
    ax2.set_ylabel('Revenue (USD)', color='green')
    ax2.tick_params(axis='y', labelcolor='green')

    # Set the title of the graph
    plt.title(f'{company_name} Stock Price and Revenue Over Time')

    # Show the graph
    fig.tight_layout()
    plt.show()

# Call the function to plot the graph for GameStop
make_graph(combined_data, 'GameStop')
````
